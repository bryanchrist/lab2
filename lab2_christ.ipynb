```python
nasa_api=os.environ['nasa_api']
useragent_url = 'https://httpbin.org/user-agent'
#website that exists to help you work with other APIs, generalized code that works in any situation, including a virtual machine
r= requests.get(useragent_url)
useragent= json.loads(r.text)['user-agent']
#standard thing to put in HTTP header to let websites know what kind of system you are using: best practice is to include it even
#though not all websites require it
```

```python
headers = {'X-API-Key': nasa_api, 
          'User-Agent': useragent, 
          'From': 'brc4cb@virginia.edu'}
#Headers are what we send API about ourselves, adds an email address so they can reach you if you abuse the API
```

```python
#Connect to API and verify connection
root = "https://api.nasa.gov"
start_date = "2022-09-24"
end_date = "2022-10-01"
endpoint = "/neo/rest/v1/feed"
params = "?start_date={START_DATE}&end_date={END_DATE}&api_key={API_KEY}".format(API_KEY= nasa_api, START_DATE= start_date, END_DATE = end_date)

r = requests.get(root + endpoint + params, 
                 headers = headers)
r
```

```python
myjson = json.loads(r.text)
nasa1 = pd.json_normalize(myjson, record_path = ['near_earth_objects', '2022-10-01'])
nasa2 = pd.json_normalize(myjson, record_path = ['near_earth_objects', '2022-09-30'])
nasa3 = pd.json_normalize(myjson, record_path = ['near_earth_objects', '2022-09-29'])
nasa4 = pd.json_normalize(myjson, record_path = ['near_earth_objects', '2022-09-28'])
nasa5 = pd.json_normalize(myjson, record_path = ['near_earth_objects', '2022-09-27'])
nasa6 = pd.json_normalize(myjson, record_path = ['near_earth_objects', '2022-09-26'])
nasa7 = pd.json_normalize(myjson, record_path = ['near_earth_objects', '2022-09-25'])
nasa = pd.concat([nasa1, nasa2, nasa3, nasa4, nasa5, nasa6, nasa7], ignore_index=True)
nasa['close_approach_date'] = [d[0]['close_approach_date']for d in nasa['close_approach_data']]
nasa
```

Question 4 Part A: Scraping house prices of Zillow is not permitted, as they have disallowed scraping for every page of their website for all user-agents with a "/r/" under disallow, which is a regex pattern that captures all possible character combinations. 

Question 4 Part B: Google does allow scraping of stock prices through Google Finance, as the following is found in an Allow section of their robots.txt file: "Allow: /finance." This would allow the user-agent to access any webpage with the finance backslash, which includes stock prices. 

Question 4 Part C: Twitter would not allow access to web scrape who NBA players follow, as the following is found in a Disallow section of their robots.txt file: "Disallow: /*/following." This would preclude anyone from scraping who any twitter user follows, as the * captures all Twitter users and the /following blocks anyone from accessing the following backslash of the URL, which is the way to access who someone is following. 

Question 4 Part D: Genius allows webscraping of their songs. There is no specific Disallow statement that has a backslash having anything to do with the word "lyrics," so any user-agent is free to scrape their lyrics.

```python
useragent_url = 'https://httpbin.org/user-agent'
#website that exists to help you work with other APIs, generalized code that works in any situation, including a virtual machine
r= requests.get(useragent_url)
useragent= json.loads(r.text)['user-agent']
#standard thing to put in HTTP header to let websites know what kind of system you are using: best practice is to include it even
#though not all websites require it
```

```python
headers = {'User-Agent': useragent, 
          'From': 'brc4cb@virginia.edu'}
#Headers are what we send API about ourselves, adds an email address so they can reach you if you abuse the API
```

```python
#Connect to API and verify connection
root = "https://maps.spirithalloween.com"
endpoint = "/api/getAsyncLocations"
lat= "40.380028"
lng = "-97.910156"
radius = "1800"
limit="1500"
params = "?template=domain&level=domain&lat={lat}&lng={lng}&radius={radius}&limit={limit}".format(lat=lat, lng=lng, radius=radius, limit=limit)
r = requests.get(root + endpoint + params, 
                 headers = headers)
r
```

```python
locations = []
for i in range(0,1500):
    try:
        myjson = json.loads(r.text)
        myjson = myjson['markers'][i]['info']
        myhtml = BeautifulSoup(myjson, 'html.parser')
        text = myhtml.find_all('div', 'address-two')[0].text
        locations.append(text)
    except:
        pass
#Gets all locations
```

```python
matches = []
for location in locations:
    if "Former" in location:
        matches.append(location)
#pulls matches
```

```python
#count matches
formerlocsDF= pd.DataFrame()
formerlocsDF['formerlocs'] = matches
formerlocsDF['formerlocs'].value_counts()
```
